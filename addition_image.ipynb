```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# ---------------------------
# Helper function: draw layer
# ---------------------------
def draw_layer(ax, y, height, color, text):
    rect = patches.Rectangle((0, y), 1, height, linewidth=1.5,
                             edgecolor='black', facecolor=color)
    ax.add_patch(rect)
    ax.text(0.5, y + height/2, text, ha='center', va='center', fontsize=12)

# ---------------------------
# Create figure
# ---------------------------
fig, ax = plt.subplots(figsize=(4, 8))

# Define layers (bottom â†’ top)
layers = [
    (0.00, 0.20, "#cfe2f3", "Insurer Retention / Deductible"),
    (0.20, 0.20, "#bdd7ee", "Quota Share\n(e.g. 30% ceded)"),
    (0.40, 0.20, "#9dc3e6", "FHCF Layer\n(State Base Layer)"),   # moved down
    (0.60, 0.20, "#6fa8dc", "XOL Layer 1\n(Working Layer)"),
    (0.80, 0.20, "#3c78d8", "XOL Layer 2\n(CAT Layer)")
]

# Draw layers
for y, h, color, text in layers:
    draw_layer(ax, y, h, color, text)

# Formatting
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

plt.title(" Reinsurance Layer Structure for CAT Risk", fontsize=14)
plt.show()
```
